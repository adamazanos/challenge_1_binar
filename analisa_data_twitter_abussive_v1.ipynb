#B.i.   DI BERESI FORMATNYA JADI KECIL SEMUA

#B.iii.   PAKAI REGEX HAPUS HAPUSIN YANG KOTOR




DARI https://www.kaggle.com/code/corneliustantius/indonesia-hatespeech-recognition-main-notebook

* sebelumnya coba buat def satu satu, tetapi tidak bisa di pakai, sekarang coba langkah b.1 b.2 langsung 1 def


https://www.kaggle.com/code/ilhamfp31/simpler-preprocess-indonesian-hate-abusive-text/notebook

```python
  def lowercase(text):
      return text.lower()

  alay_dict_map = dict(zip(alay_dict['original'], alay_dict['replacement']))
  def normalize_alay(text):
      return ' '.join([alay_dict_map[word] if word in alay_dict_map else word for word in text.split(' ')])

  def remove_nonaplhanumeric(text):
    text = re.sub('[^0-9a-zA-Z]+', ' ', text) 
    return text
  
  def remove_unnecessary_char(text):
    text = re.sub('\n',' ',text) # Remove every '\n'
    text = re.sub('\n\n',' ',text) # Remove every '\n'
    text = re.sub('rt',' ',text) # Remove every retweet symbol
    text = re.sub('user',' ',text) # Remove every username
    text = re.sub('((www\.[^\s]+)|(https?://[^\s]+)|(http?://[^\s]+))',' ',text) # Remove every URL
    text = re.sub('  +', ' ', text) # Remove extra spaces
    text = re.sub('x[0-9a-zA-Z]{2}',' ',text) # buang emoji utf-8 thanks gaz
    return text


uji_v1 = "BEBAAss lah, BABE. bagus ga nih BUAT nG3Tesnya " #BUAT CONTOH KALIMAT

uji_v2 = "aamiin adek abis" #BUAT CONTOH KALIMAT

uji_v3 = "Hehe\n\n rt user user user apa kabs www.google.com\n  hehe" #BUAT CONTOH KALIMAT

uji_v4 = "Halooo,,,,, duniaa!!" #BUAT CONTOH KALIMAT

uji_v5 = " xf0 x9f x98 x82 hoaah" #BUAT CONTOH KALIMAT


print("SEBELUM lowercase: ",  (uji_v1), " ==>> " "SESUDAH lowercase: ", lowercase(uji_v1))

print("SEBELUM normalize_alay: ",  (uji_v2), " ==>> " "SESUDAH normalize_alay: ", normalize_alay(uji_v2))

print("SEBELUM remove_unnecessary_char: ",  (uji_v3), " ==>> " "SESUDAH remove_unnecessary_char: ", remove_unnecessary_char(uji_v3))

print("SEBELUM remove_nonaplhanumeric: ",  (uji_v4), " ==>>" "SESUDAH remove_nonaplhanumeric ", remove_nonaplhanumeric(uji_v4))

print("SEBELUM remove_unnecessary_char: ",  (uji_v5), " ==>>" "SESUDAH remove_unnecessary_char ", remove_unnecessary_char(uji_v5))
```

```python
def preprocess(text):
    text = lowercase(text) # 1
    text = remove_unnecessary_char(text) # 2
    text = normalize_alay(text) # 3
    text = remove_nonaplhanumeric(text)
    return text

uji_v4 = "BAAss lah, BABE. aamiin xf0 x9f x98 x82 adek \n rt user user apa kabss www.google.com\n  hehe"

print("SEBELUM preprocess: ",  (uji_v4), " ==>> " "SESUDAH preprocess: ", preprocess(uji_v4))
```
